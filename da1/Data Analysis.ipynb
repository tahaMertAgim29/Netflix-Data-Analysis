Movie vs TV Show Visualization using Seaborn and Matplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='type')
plt.title("Movie vs TV Show")
plt.show()
```

Top Country Content and Countries

```python
df['country'].value_counts().head(10) # Shows 10 rows


top_country = df['country'].value_counts().head(15)

top_country.plot(kind='barh')
plt.title("Top Fifteen Countries")
plt.show()
```

Missing Value Visualization

```python
sns.heatmap(df.isnull(), cbar=False, cmap='coolwarm')
plt.title("Missing Values")
plt.show()
```

Rating Analysis

```python
top_ratings = df['rating'].value_counts().head(10)


top_ratings.plot(kind='pie')
plt.title("Top Ten Ratings")
plt.show()

```

Year Analysis

```python
top_years = df['release_year'].value_counts().head(10)


top_years.plot(kind='bar')

plt.title("Top 10 Release Years")
plt.show()

```

Most & Max Analysis

```python
summary = {
    'Top Country': df['country'].value_counts().idxmax(),
    'The Most Common Content Type': df['type'].value_counts().idxmax(),
    'The Most Common Rating': df['rating'].value_counts().idxmax(),
    'The Most Common Genre': df['listed_in'].value_counts().idxmax(),
}

print(summary)
```

Genre Analysis

```python
top_genre = df['listed_in'].value_counts().head(10).plot(kind='barh')
plt.title("Top Ten Genres")
plt.show()
```

```python
# The Most Director
df['director'].value_counts().idxmax()


# Num of Documentaries grouped by Country
df['country'][df['listed_in'] == 'Documentaries'].value_counts()



```
